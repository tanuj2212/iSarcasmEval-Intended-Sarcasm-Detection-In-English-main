```python
import nlpaug.augmenter.char as nac
import nlpaug.augmenter.word as naw
import nlpaug.augmenter.sentence as nas
import nlpaug.flow as naf
import pandas as pd

from nlpaug.util import Action
```

```python
df=pd.read_csv('train.En.csv')
df.head(1)
```

```python
df1=df[:868]
```

```python
TOPK=20 #default=100
ACT = 'insert' #"substitute"
 
aug_bert = naw.ContextualWordEmbsAug(
    model_path='distilbert-base-uncased', 
    #device='cuda',
    action=ACT, top_k=TOPK)
# print("Original:")
# print(test_sentence)
# print("Augmented Text:")
```

```python
d={}
a =0
for tweet in list(df1['tweet'][0:867]):
  #if list(df['irony']==1):
  #if  df1.loc[df1[df1['tweet']==tweet].index.values[0],'sarcastic']==1:
    for i in range(3):
      augmented_text = aug_bert.augment(tweet)
      d[augmented_text]=1
    a=a+1
    print(a)
```

```python
len(d)
```

```python
data = {'tweet':list(d.keys()),
        'sarcastic': list(d.values())} 
new = pd.DataFrame.from_dict(data)
  
new.head(5)
```

```python
frames = [new, df]
  
result = pd.concat(frames)
```
